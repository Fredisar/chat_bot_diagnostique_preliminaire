# Importation et Prémière exploration

```python
# Importation des bibliothèques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuration de l'affichage
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
# Chargement du dataset
df = pd.read_csv('Training.csv')
```

```python
# 1. Aperçu des données
print("=" * 80)
print("📊 APERÇU DES DONNÉES")
print("=" * 80)
print(f"Nombre de lignes : {df.shape[0]}")
print(f"Nombre de colonnes : {df.shape[1]}")
print("\nPremières 5 lignes :")
print(df.head())
```

```python
# 2. Informations générales
print("\n" + "=" * 80)
print("📋 INFORMATIONS GÉNÉRALES")
print("=" * 80)
print(df.info())
```

```python
# 3. Statistiques descriptives
print("\n" + "=" * 80)
print("📈 STATISTIQUES DESCRIPTIVES")
print("=" * 80)
print(df.describe())
```

# 🔍 Étape 2 : Analyse des colonnes et des types

```python
# 4. Liste des colonnes
print("=" * 80)
print("📑 LISTE DES COLONNES")
print("=" * 80)
columns = df.columns.tolist()
print(f"Nombre total de colonnes : {len(columns)}")
print("\nPremières 20 colonnes :")
for i, col in enumerate(columns[:20], 1):
    print(f"{i:3d}. {col}")
```

```python
# 5. Vérification des types de données
print("\n" + "=" * 80)
print("🔍 TYPES DE DONNÉES")
print("=" * 80)
print(df.dtypes.value_counts())
```

```python
# 6. Vérification des valeurs manquantes
print("\n" + "=" * 80)
print("❓ VALEURS MANQUANTES")
print("=" * 80)
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({
    'Valeurs manquantes': missing_values,
    'Pourcentage': missing_percentage
})
print(missing_df[missing_df['Valeurs manquantes'] > 0])
print(f"\nNombre total de valeurs manquantes : {df.isnull().sum().sum()}")
```

```python
# 7. Suppression des colonnes inutiles (si présentes)
# Vérifier si une colonne 'Unnamed' existe
unnamed_cols = [col for col in df.columns if 'Unnamed' in col]
if unnamed_cols:
    print(f"Suppression des colonnes : {unnamed_cols}")
    df = df.drop(columns=unnamed_cols)
```

```python
# 8. Vérification des doublons
print("\n" + "=" * 80)
print("🔄 DOUBLONS")
print("=" * 80)
duplicates = df.duplicated().sum()
print(f"Nombre de lignes dupliquées : {duplicates}")
```

```python
# 9. Suppression des doublons
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"✅ {duplicates} doublons supprimés")
    print(f"Nouveau nombre de lignes : {df.shape[0]}")
```

```python
# 10. Vérification des valeurs aberrantes dans les colonnes de symptômes
print("\n" + "=" * 80)
print("⚠️ VALEURS ABERRANTES")
print("=" * 80)
```

```python
# Analyser les colonnes de symptômes (sauf la dernière colonne 'prognosis')
symptom_cols = df.columns[:-1]
for col in symptom_cols[:10]:  # On vérifie les 10 premières
    unique_values = df[col].unique()
    print(f"{col}: {sorted(unique_values)}")
```

```python
# 11. Vérification que toutes les colonnes de symptômes sont binaires (0 ou 1)
print("\n" + "=" * 80)
print("✅ VÉRIFICATION DES VALEURS BINAIRES")
print("=" * 80)
invalid_cols = []
for col in symptom_cols:
    if not set(df[col].unique()).issubset({0, 1}):
        invalid_cols.append(col)
        print(f"⚠️ {col} contient des valeurs non-binaires : {df[col].unique()}")

if not invalid_cols:
    print("✅ Toutes les colonnes de symptômes sont correctement encodées (0 ou 1)")
```

```python
# 12. Analyse des maladies (prognosis)
print("=" * 80)
print("🏥 ANALYSE DES MALADIES")
print("=" * 80)

# Distribution des maladies
disease_counts = df['prognosis'].value_counts()
print(f"Nombre de maladies uniques : {len(disease_counts)}")
print(f"\nTop 10 des maladies les plus fréquentes :")
print(disease_counts.head(10))

# Visualisation de la distribution
plt.figure(figsize=(12, 8))
disease_counts.head(20).plot(kind='bar')
plt.title('Top 20 des maladies les plus fréquentes', fontsize=16)
plt.xlabel('Maladies', fontsize=12)
plt.ylabel('Nombre de cas', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 13. Équilibre des classes
print("\n" + "=" * 80)
print("⚖️ ÉQUILIBRE DES CLASSES")
print("=" * 80)
min_freq = disease_counts.min()
max_freq = disease_counts.max()
mean_freq = disease_counts.mean()
print(f"Fréquence minimale : {min_freq}")
print(f"Fréquence maximale : {max_freq}")
print(f"Fréquence moyenne : {mean_freq:.2f}")
print(f"Rapport max/min : {max_freq/min_freq:.2f}")
```

# 🔗 Étape 5 : Analyse des corrélations

```python
# 14. Analyse de corrélation (sur un échantillon pour éviter les problèmes de mémoire)
print("=" * 80)
print("📊 CORRÉLATIONS")
print("=" * 80)
# Sélectionner un sous-ensemble de symptômes pour la visualisation
sample_symptoms = symptom_cols[:20]  # Premiers 20 symptômes
correlation_matrix = df[sample_symptoms].corr()

# Visualisation de la matrice de corrélation
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm',
            cbar_kws={'label': 'Corrélation'},
            square=True)
plt.title('Matrice de corrélation des symptômes (20 premiers)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 15. Symptômes les plus fréquents
print("\n" + "=" * 80)
print("📋 SYMPTÔMES LES PLUS FRÉQUENTS")
print("=" * 80)
symptom_freq = df[symptom_cols].sum().sort_values(ascending=False)
print("Top 20 des symptômes les plus fréquents :")
print(symptom_freq.head(20))

# Visualisation
plt.figure(figsize=(12, 8))
symptom_freq.head(20).plot(kind='bar', color='skyblue')
plt.title('Top 20 des symptômes les plus fréquents', fontsize=16)
plt.xlabel('Symptômes', fontsize=12)
plt.ylabel('Fréquence', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 🎯 Étape 6 : Analyse par maladie

```python
# 16. Analyse des symptômes par maladie
print("=" * 80)
print("🔬 SYMPTÔMES PAR MALADIE")
print("=" * 80)

# Choisir une maladie spécifique pour analyse
selected_disease = disease_counts.index[0]  # La plus fréquente
print(f"Analyse pour : {selected_disease}")

# Filtrer les données pour cette maladie
disease_data = df[df['prognosis'] == selected_disease]
symptom_freq_disease = disease_data[symptom_cols].sum().sort_values(ascending=False)

print(f"\nTop 10 des symptômes pour {selected_disease} :")
print(symptom_freq_disease.head(10))
```

```python
# 17. Comparaison de deux maladies
print("\n" + "=" * 80)
print("🔄 COMPARAISON DE DEUX MALADIES")
print("=" * 80)

disease1 = disease_counts.index[0]
disease2 = disease_counts.index[1] if len(disease_counts) > 1 else disease_counts.index[0]

print(f"Comparaison : {disease1} vs {disease2}")

# Calculer les fréquences des symptômes pour chaque maladie
freq1 = df[df['prognosis'] == disease1][symptom_cols].mean()
freq2 = df[df['prognosis'] == disease2][symptom_cols].mean()

# Afficher les différences
diff = freq1 - freq2
print(f"\nSymptômes qui différencient {disease1} de {disease2} :")
print("Top 5 symptômes plus fréquents dans", disease1)
print(diff.sort_values(ascending=False).head(5))
print("\nTop 5 symptômes plus fréquents dans", disease2)
print(diff.sort_values(ascending=True).head(5))
```

# 💾 Étape 7 : Sauvegarde du dataset nettoyé

```python
# 18. Sauvegarde du dataset nettoyé
print("=" * 80)
print("💾 SAUVEGARDE")
print("=" * 80)

# Sauvegarder le dataset nettoyé
df.to_csv('training_cleaned.csv', index=False)
print("✅ Dataset nettoyé sauvegardé dans 'data/training_cleaned.csv'")
```

```python
# 19. Statistiques finales
print("\n" + "=" * 80)
print("📊 STATISTIQUES FINALES")
print("=" * 80)
print(f"Nombre total de lignes : {df.shape[0]}")
print(f"Nombre total de colonnes : {df.shape[1]}")
print(f"Nombre de symptômes : {len(symptom_cols)}")
print(f"Nombre de maladies uniques : {len(df['prognosis'].unique())}")
print(f"Taux de remplissage : {100 - (df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100):.2f}%")
```

# 📝 Étape 8 : Rapport d'analyse

```python
# 20. Génération d'un rapport d'analyse
print("=" * 80)
print("📄 RAPPORT D'ANALYSE")
print("=" * 80)

report = f"""
RAPPORT D'ANALYSE DES DONNÉES MÉDICALES
========================================

1. DIMENSIONS DU DATASET
   - Lignes : {df.shape[0]}
   - Colonnes : {df.shape[1]}
   - Symptômes : {len(symptom_cols)}
   - Maladies : {len(df['prognosis'].unique())}

2. QUALITÉ DES DONNÉES
   - Valeurs manquantes : {df.isnull().sum().sum()}
   - Doublons : {duplicates}
   - Taux de remplissage : {100 - (df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100):.2f}%

3. DISTRIBUTION
   - Maladie la plus fréquente : {disease_counts.index[0]} ({disease_counts.max()} cas)
   - Maladie la moins fréquente : {disease_counts.index[-1]} ({disease_counts.min()} cas)
   - Équilibre des classes : {'Oui' if max_freq/min_freq < 2 else 'Non'}

4. SYMPTÔMES
   - Symptôme le plus fréquent : {symptom_freq.index[0]} ({symptom_freq.max()} occurrences)
   - Symptôme le moins fréquent : {symptom_freq.index[-1]} ({symptom_freq.min()} occurrences)
   - Nombre moyen de symptômes par maladie : {df[symptom_cols].sum(axis=1).mean():.2f}
   - Nombre maximum de symptômes pour une maladie : {df[symptom_cols].sum(axis=1).max()}
   - Nombre minimum de symptômes pour une maladie : {df[symptom_cols].sum(axis=1).min()}

5. RECOMMANDATIONS
   - {'✅' if invalid_cols == [] else '⚠️'} Les données sont correctement encodées
   - {'⚠️' if duplicates > 0 else '✅'} Aucun doublon détecté
   - {'⚠️' if df.isnull().sum().sum() > 0 else '✅'} Pas de valeurs manquantes
"""

print(report)

# Sauvegarder le rapport
with open('analysis_report.txt', 'w', encoding='utf-8') as f:
    f.write(report)
print("\n✅ Rapport sauvegardé dans 'data/analysis_report.txt'")
```
